```python
def petershagen_lock(folder_path):
    for filename in os.listdir(folder_path):
        if (filename.startswith("128-43") or filename.startswith("44-43")) and filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            
            # Convert ais_timestamp to datetime
            df['ais_timestamp'] = pd.to_datetime(df['ais_timestamp'])

            # Function to calculate remaining time and set lock_status
            def calculate_remaining_time(row):
                timestamp = row['ais_timestamp']
                weekday = timestamp.weekday()
                hour = timestamp.hour
                minute = timestamp.minute
                remaining_time = 0
                lock_status = 1

                # For Monday to Saturday between 21:30 and 6:00
                if timestamp.date() >= datetime.date(2023, 1, 1) and timestamp.date() <= datetime.date(2023, 12, 31):
                    if weekday < 6 and (hour > 21 or (hour == 21 and minute >= 30) or hour < 6):
                        remaining_time = ((6 - hour) % 24) + (0 - minute) / 60
                        lock_status = 0

                # For Sunday between 16:30 and 9:00
                if timestamp.date() >= datetime.date(2023, 7, 2) and timestamp.date() <= datetime.date(2023, 12, 31):
                    if weekday == 6 and (hour > 16 or (hour == 16 and minute >= 30) or hour < 9):
                        remaining_time = ((9 - hour) % 24) + (0 - minute) / 60
                        lock_status = 0

                row['travel_time_duration'] += remaining_time
                row['lock_status'] = lock_status
                return row

            # Apply the function to each row
            df = df.apply(calculate_remaining_time, axis=1)
            
            # Save the modified DataFrame back to CSV
            df.to_csv(file_path, index=False)
```
