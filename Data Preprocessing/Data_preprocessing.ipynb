```python
# Function to adjust "24:00" to "00:00" of the next day
def adjust_datetime(row):
    if row['time'] == '24:00':
        # Convert to datetime, add a day, and format
        new_date = pd.to_datetime(row['date'], dayfirst=True) + pd.Timedelta(days=1)
        return new_date.strftime('%d.%m.%Y') + ' 00:00'
    else:
        # Combine date and time normally
        return row['date'] + ' ' + row['time']
```

```python
def safe_convert_timestamp(column, output_format='%d.%m.%Y %H:%M'):
    """
    Safely converts a timestamp column to the desired format, handling multiple input formats.

    Parameters:
    column (pd.Series): The timestamp column to convert.
    output_format (str): The desired output format for the timestamps.

    Returns:
    pd.Series: The converted timestamp column.
    """
    converted = pd.Series(index=column.index, dtype='object')

    for i, timestamp in column.items():
        try:
            # Try the first format
            converted[i] = pd.to_datetime(timestamp, format='%Y-%m-%d %H:%M:%S').strftime(output_format)
        except ValueError:
            try:
                # Try the second format
                converted[i] = pd.to_datetime(timestamp, format='%d.%m.%Y %H:%M').strftime(output_format)
            except ValueError:
                # Handle other cases, e.g., NaNs or different format
                converted[i] = pd.NaT  # Assign a not-a-time value, or handle as you see fit

    return converted
```

```python
# Convert 'max_cm' to a numeric data type, coercing any non-numeric values to NaN
main_data['max_cm']   = pd.to_numeric(main_data['max_cm'], errors='coerce')
main_data['sig_cm']   = pd.to_numeric(main_data['sig_cm'], errors='coerce')
main_data['turb_cm']  = pd.to_numeric(main_data['turb_cm'], errors='coerce')
main_data['tp']       = pd.to_numeric(main_data['tp'], errors='coerce')
```

```python
# Fill missing values
main_data['max_cm'].fillna(method='ffill', inplace=True)
main_data['sig_cm'].fillna(method='ffill', inplace=True)
main_data['turb_cm'].fillna(method='ffill', inplace=True)
main_data['tp'].fillna(method='ffill', inplace=True)
main_data['water_level_cm'].fillna(method='ffill', inplace=True)
main_data['lock_status'].fillna(method='ffill', inplace=True)
```

```python
# Fill missing values of 'heading' with the mode
main_data['heading'].fillna(main_data['heading'].mode()[0], inplace=True)
```

```python
# Convert degrees to radians
main_data['radians'] = np.radians(main_data['heading'])

# Calculate sine and cosine
main_data['sine']   = np.sin(main_data['radians'])
main_data['cosine'] = np.cos(main_data['radians'])

#paper reference
```

```python
# Calculate Q1 and Q3
Q1 = main_data['travel_time_duration'].quantile(0.25)
Q3 = main_data['travel_time_duration'].quantile(0.75)

# Calculate Interquartile Range (IQR)
IQR = Q3 - Q1

# Determine the outliers boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove the outliers
main_data = main_data[(main_data['travel_time_duration'] >= lower_bound) & (main_data['travel_time_duration'] <= upper_bound)]
```
