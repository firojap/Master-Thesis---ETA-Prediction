```python
def display_map(df, lat_column="lat", lon_column="lon", zoom_level=10):
    """
    Displays a map with data points from a DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data points.
        lat_column (str): The name of the latitude column in df. Default is 'lat'.
        lon_column (str): The name of the longitude column in df. Default is 'lon'.
        zoom_level (int): The initial zoom level of the map. Default is 10.
    """
    # Calculate the mean of the latitude and longitude for the center of the map
    center_lat = df[lat_column].mean()
    center_lon = df[lon_column].mean()

    fig = px.scatter_mapbox(df, lat=lat_column, lon=lon_column, zoom=zoom_level,
                            center={"lat": center_lat, "lon": center_lon})
    fig.update_layout(mapbox_style="open-street-map", margin={"r":0, "t":0, "l":0, "b":0})
    fig.show()
```

```python
def plot_distance_over_time(df):
    """
    Plots distance over time from a given DataFrame.

    Parameters:
    df (DataFrame): A pandas DataFrame containing 'ais_timestamp' and 'distance' columns.
    """
    # Ensure ais_timestamp is in datetime format
    if not pd.api.types.is_datetime64_any_dtype(df['ais_timestamp']):
        df['ais_timestamp'] = pd.to_datetime(df['ais_timestamp'])

    # Plotting
    plt.figure(figsize=(10, 6))  # Adjust the size of the plot
    plt.plot(df['ais_timestamp'], df['distance'], marker='o')  # 'o' is for circular markers
    plt.xlabel('Timestamp')
    plt.ylabel('Distance (km)')
    plt.title('Distance Over Time')
    plt.xticks(rotation=45)  # Rotates the x-axis labels for better readability
    plt.tight_layout()  # Adjusts the plot to ensure everything fits without overlapping
    plt.show()
```
